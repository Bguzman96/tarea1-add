```python
print(df_sii.groupby("mes", as_index = True).agg("mean"))
```

```python
print(df_sii.groupby(['year','mes'], as_index = False).agg("mean"))
```

Al utilizar el argumento "as_index = True" y a su vez agrupando solo por mes toma el valor promedio de los años 2019 y 2020 (por el uso de la función .mean). En cambio al agrupar por año y "as_index = False" considera los meses independientes del año, calculando un mes por cada año dentro del DF.

3. ¿Qué podrías inferir de los missing values y los días hábiles? Úsalos para crear una nueva variable llamada dia_de_la_semana en df_sii con valores "Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo".

```python
df_sii
```

Los *missing values* se presentan en los días que no son hábiles (feriados y fines de semana) o no existen en el calendario, como por ejemplo algunos días de febrero.

```python
months = {
    "Ene": 1,
    "Feb": 2,
    "Mar": 3,
    "Abr": 4,
    "May": 5,
    "Jun": 6,
    "Jul": 7,
    "Ago": 8,
    "Sep": 9,
    "Oct": 10,
    "Nov": 11,
    "Dic": 12,
}

days = {
    "Monday": "Lunes",
    "Tuesday": "Martes",
    "Wednesday": "Miércoles",
    "Thursday": "Jueves",
    "Friday": "Viernes",
    "Saturday": "Sábado",
    "Sunday": "Domingo"
}

dias = []

for row in range(0, len(df_sii)):
    try:
        dias.append(days[datetime.datetime(df_sii["year"][row], months[df_sii["mes"][row]], df_sii["Día"][row]).strftime("%A")])
    except:
        dias.append(np.nan)

dias = pd.Series(dias)
df_sii["dia_de_la_semana"] = dias
```

```python
hoja_80 = pd.read_excel("../data/raw/dolar_observado_bc.xlsx", skiprows = 2, dtype = {'Periodo': str}, sheet_name = "80")
hoja_90 = pd.read_excel("../data/raw/dolar_observado_bc.xlsx", skiprows = 2, dtype = {'Periodo': str}, sheet_name = "90")
hoja_00 = pd.read_excel("../data/raw/dolar_observado_bc.xlsx", skiprows = 2, dtype = {'Periodo': str}, sheet_name = "00")
hoja_10 = pd.read_excel("../data/raw/dolar_observado_bc.xlsx", skiprows = 2, dtype = {'Periodo': str}, sheet_name = "10")
hoja_20 = pd.read_excel("../data/raw/dolar_observado_bc.xlsx", skiprows = 2, dtype = {'Periodo': str}, sheet_name = "20")
```

```python
hoja_80.head(15)
```

```python
def agrega_columnas(df):
    dias = []
    meses = []
    years = []

    for row in range(0, len(df)):
        datos = df["Periodo"][row].split(" ")[0].split("-")
        dias.append(datos[2])
        meses.append(datos[1])
        years.append(datos[0])

    dias = pd.Series(dias)
    meses = pd.Series(meses)
    years = pd.Series(years)

    df["dia"] = dias
    df["mes"] = meses
    df["year"] = years
    return df
```

```python
hoja_80 = agrega_columnas(hoja_80)
hoja_90 = agrega_columnas(hoja_90)
hoja_00 = agrega_columnas(hoja_00)
hoja_10 = agrega_columnas(hoja_10)
hoja_20 = agrega_columnas(hoja_20)
```

```python
df_bc_all = pd.concat([hoja_80, hoja_90, hoja_00, hoja_10, hoja_20]) #Unimos ambos datasets en uno solo
```
