Ocupa el método str.split para crear las columnas dia,mes y year en cada dataframe.

```python
def agrega_columnas(df):
    dias = []
    meses = []
    years = []

    for row in range(0, len(df)):
        datos = df["Periodo"][row].split(" ")[0].split("-")
        dias.append(datos[2])
        meses.append(datos[1])
        years.append(datos[0])

    dias = pd.Series(dias)
    meses = pd.Series(meses)
    years = pd.Series(years)

    df["dia"] = dias
    df["mes"] = meses
    df["year"] = years
    return df
```

```python
hoja_80 = agrega_columnas(hoja_80)
hoja_90 = agrega_columnas(hoja_90)
hoja_00 = agrega_columnas(hoja_00)
hoja_10 = agrega_columnas(hoja_10)
hoja_20 = agrega_columnas(hoja_20)
```

```python
df_bc_all = pd.concat([hoja_80, hoja_90, hoja_00, hoja_10, hoja_20]) #Unimos ambos datasets en uno solo
```

```python
df_merged = pd.merge(df_bc_all_tidy, df_sii_tidy, on=['dia', 'mes', 'year'], how='right')
df_merged
```

Crear variable [día, mes]

```python
dia_mes = []
for row in range(0, len(df_merged)):
    if (df_merged["dia_de_la_semana"][row] != "Sábado") & (df_merged["dia_de_la_semana"][row] != "Domingo") & (np.isnan(df_merged["1.Dólar observado_x"][row])) & (~pd.isna(df_merged["dia_de_la_semana"][row])):    
        dia_mes.append(np.array([df_merged["dia"][row], df_merged["mes"][row]]).tolist())
```

Buscar valores nulos para esas combinaciones de días y meses en el DataFrame general df_bc_all.

```python
results = pd.DataFrame(
    {"dia_mes": dia_mes}
)

count = []

for element in dia_mes:
    count.append(len(df_bc_all[(df_bc_all["dia"] == element[0]) & (df_bc_all["mes"] == element[1]) & (np.isnan(df_bc_all["1.Dólar observado"]))]))

count = pd.Series(count)
results["count"] = count
results
```

Se puede inferir que las combinaciones *[dia, mes]* con mayor cantidad de NaN son días feriados.
